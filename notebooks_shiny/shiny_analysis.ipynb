```python
import json
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
```

```python
conuv="/home/aalbiol/rasp4/mscan_dvc/dvc_s90_orange/out_evaluate_shiny_con_UV/test_scores.json"
sinuv="/home/aalbiol/rasp4/mscan_dvc/dvc_s90_orange/out_evaluate_shiny_sin_UV/test_scores.json"

with open(conuv) as f:
    conuv_scores = json.load(f)
with open(sinuv) as f:
    sinuv_scores = json.load(f)
```

```python
con=conuv_scores['test_results']
sin=sinuv_scores['test_results']
```

```python
scores_con_positive = []
scores_sin_positive = []
scores_con_negative = []
scores_sin_negative = []

names_con_positive = []
names_sin_positive = []
names_con_negative = []
names_sin_negative = []
for caso in con:
    scores=caso['scores']
    gt=caso['ground_truth']
    gt_rotten=int(gt['rotten'])
    name=caso['filename']
    score_rotten=scores['rotten']
    if gt_rotten==1:
        scores_con_positive.append(score_rotten)
        names_con_positive.append(name)
        
    else:
        scores_con_negative.append(score_rotten)
        names_con_negative.append(name)
        
for caso in sin:
    scores=caso['scores']
    gt=caso['ground_truth']
    gt_rotten=int(gt['rotten'])
    score_rotten=scores['rotten']
    name=caso['filename']
    if gt_rotten==1:
        scores_sin_positive.append(score_rotten)
        names_sin_positive.append(name)
        
    else:
        scores_sin_negative.append(score_rotten)
        names_sin_negative.append(name)
```

```python
print(len(scores_con_positive), len(scores_sin_positive))
print(len(scores_con_negative), len(scores_sin_negative))
```

```python
for c,s in zip(names_con_positive, names_sin_positive):
    if c != s:
        print("ERROR: names do not match", c, s)
```

```python
_=plt.plot(scores_con_positive, label='con UV')
_=plt.plot(scores_sin_positive, label='sin UV')
_=plt.grid()
_=plt.legend()
_=plt.ylim(0, 1)
_=plt.title('Scores for rotten fruits (positive cases)')
```

```python
_=plt.plot(scores_con_negative, label='con UV')
_=plt.plot(scores_sin_negative, label='sin UV')
_=plt.grid()
_=plt.legend()
_=plt.ylim(0, 1)
_=plt.title('Scores for non-rotten fruits (negative cases)')
```
