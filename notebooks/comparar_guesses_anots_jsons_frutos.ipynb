```python
import sys
import os
import shutil

from glob import glob
from tqdm import tqdm
import json
import numpy as np  
```

```python
guessdir="/home/aalbiol/Desktop/Guesses_Alba_5_y_6_Sara_jsons_only/imgs_Sara_guess"
anotdir="/home/aalbiol/Downloads/imgSara_anotadas"
```

```python
guesses=glob(guessdir+"/*.json")
anots=glob(anotdir+"/*.json")
print("Number of guesses: ", len(guesses))
print("Number of annotations: ", len(anots))
```

```python
fruits=[]
number_of_labels=0
diferencias={}
positivos_guess={}
positivos_anot={}
for anot in tqdm(anots):
    anotbasebame=os.path.basename(anot)
    
    for guess in guesses:
        if not anotbasebame in guess:
            continue

        with open(guess) as f:
            guess_json=json.load(f)
        with open(anot) as f:
            anot_json=json.load(f)
        guess_anots=guess_json["annotations"]
        anot_anots=anot_json["annotations"]

        for key in guess_anots.keys():
            if not key in anot_anots.keys():
                continue
            guess_label=guess_anots[key]
            anot_label=anot_anots[key]

            if anot_label != guess_label:
                if key not in diferencias:
                    diferencias[key] = 1 
                else:
                    diferencias[key] += 1
            guess_label=int(guess_label)
            anot_label=int(anot_label)

            if guess_label >0:
                print(key, guess_label, anot_label)
                if key not in positivos_guess:
                    positivos_guess[key] = 1
                else:
                    positivos_guess[key] += 1
            if anot_label>0:
                if key not in positivos_anot:
                    positivos_anot[key] = 1
                else:
                    positivos_anot[key] += 1                


                                       
                
                    
           

        
```
