```python
import sys
import os
import shutil

from glob import glob
from tqdm import tqdm
import json
import numpy as np  
```

```python
srcdir="/home/aalbiol/Desktop/out"
targetdir="../data/"
```

```python
sourcejsons=glob(srcdir+"/*.json")
print(len(sourcejsons))
```

```python
# recursive search of jsons
targetjsons1=glob(targetdir+"/mandarins/"+"**/*.json", recursive=True)
targetjsons2=glob(targetdir+"/oranges/"+"**/*.json", recursive=True)
targetjsons=targetjsons1+targetjsons2
print(len(targetjsons))
```

```python
fruits=[]
number_of_labels=0
for jsonfile in tqdm(sourcejsons):
    jsonname=os.path.basename(jsonfile)
    

    print(">>>>jsonname:", jsonname)
    for f in targetjsons:
        if jsonname in f:

            with open(jsonfile, 'r') as source:
                newdata=json.load(source)
            with open(f, 'r') as target:
                olddata=json.load(target)
            new_anotations=newdata["annotations"]
            old_anotations=olddata["annotations"]

            for k,v in new_anotations.items():
                if not k in old_anotations:
                    continue
                if v != old_anotations[k]:
                    fruits.append(jsonname)
                    number_of_labels+=1
                    print("jsonname:", jsonname,"Key: ", k)
                    print("New value: ", v)
                    print("Old value: ", old_anotations[k])
fruits_set=set(fruits)
print("Number of labels:",len(fruits))
print("Number of labels: ", number_of_labels)
print("Number of different jsons: ", len(fruits_set))

                                       
                
                    
           

        
```
