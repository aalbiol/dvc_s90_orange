```python
import sys
import os
import shutil

from glob import glob
from tqdm import tqdm
import json
import numpy as np  
```

# Objetivo del cuaderno

A partir del resultado de las anotaciones de CVAT

* Carpeta con jsons + una carpeta de *"discarded"*

Imbricar lo jsons en su sitio
* Eliminar lo de discarded
* Eliminar anotaciones de vista
* copiar json al lugar de destino

```python
srcdir=r"C:\Users\csanchis\Downloads\task98_processed\task98_processed"
targetdir="../data/"
```

```python
sourcejsons=glob(srcdir+"/*.json")
print(len(sourcejsons))
```

```python
# recursive search of jsons
targetjsons1=glob(targetdir+"/5_secuencias_alba/"+"**/*.json", recursive=True)
targetjsons2=[]
targetjsons=targetjsons1+targetjsons2
print(len(targetjsons))
```

```python
def sort_dict(d):
    return {k: d[k] for k in sorted(d.keys())}  
```

```python
fruits=[]
number_of_labels=0
for jsonfile in tqdm(sourcejsons):
    jsonname=os.path.basename(jsonfile)
    

    print(">>>>jsonname:", jsonname)
    for f in targetjsons:
        if jsonname in f:

            with open(jsonfile, 'r') as source:
                newdata=json.load(source)
            with open(f, 'r') as target:
                olddata=json.load(target)
            new_anotations=newdata["annotations"]
            old_anotations=olddata["annotations"]

            for k,v in new_anotations.items():
                if not k in old_anotations:
                    continue
                if v != old_anotations[k]:
                    fruits.append(jsonname)
                    number_of_labels+=1

fruits_set=set(fruits)
print("Number of labels:",len(fruits))
print("Number of labels: ", number_of_labels)
print("Number of different jsons: ", len(fruits_set))

                                       
                
                    
           

        
```

```python
# Reemplazo
fruits=[]
number_of_labels=0
for jsonfile in tqdm(sourcejsons): # origen
    jsonname=os.path.basename(jsonfile)
    

    print(">>>>jsonname:", jsonname)
    for targetjson in targetjsons:
        if jsonname in targetjson: # Destino
            with open(jsonfile, 'r') as source:
                newdata=json.load(source)
            newdata['annotations'] = sort_dict(newdata['annotations'])
            if "views_annotations" in newdata:
               del newdata["views_annotations"]

            with open(targetjson, 'w') as target:
                json.dump(newdata, target, indent=4)

```

```python
srcdir=r"C:\Users\csanchis\Downloads\task98_processed\task98_processed\discarded"
targetdir="../data"
```

```python
sourcejsons=glob(srcdir+"/*.json")
print(len(sourcejsons))
```

```python
targetjsons1=glob(targetdir+"/5_secuencias_alba/"+"**/*.json", recursive=True)
targetjsons2=[]
targetjsons=targetjsons1+targetjsons2
print(len(targetjsons))
```

```python
for jsonfile in tqdm(sourcejsons):
    jsonname=os.path.basename(jsonfile)
    
    found=False

    for f in targetjsons:
        if jsonname in f:
            found=True
            clave=f.replace(".json","")
            lista=glob(clave+"*")
            for file in lista:
                print("Deleting file:", file)
                os.remove(file)
    if not found:
        print("Not found json:", jsonname)
```
